```python
#import dependencies
import csv
import pandas as pd
import numpy as np
import datetime as dt
```

```python
#import data as pandas dataframes
tx_tests_county = pd.read_excel("../resources/data/tx_tests_county.xlsx")
tx_hospitalizations = pd.read_excel("../resources/data/tx_hospitalizations.xlsx")
tx_fatality_county = pd.read_excel("../resources/data/tx_fatality_county.xlsx")
tx_cases_county = pd.read_excel("../resources/data/tx_cases_county.xlsx")
tx_activecases_county = pd.read_excel("../resources/data/tx_activecases_county.xlsx")
```

```python
#for later
months = ["January", "February", "March", "April", "May", "June", "July", "August", "September", "October", "November", "December"]
```

# Cleaning the Data

```python
#City Hospitalizations DF cleanup
hosp_clean = tx_hospitalizations.iloc[1:,:]
headers = hosp_clean.iloc[0]
#change datetimes into strings
for x in range(len(headers)):
    try:
        headers[x] = headers[x].strftime("%B %d")
    except Exception as e:
        pass
hosp_clean.iloc[0] = headers
hosp_clean.iloc[0,1] = "City"
hosp_clean.columns = hosp_clean.iloc[0]
hosp = hosp_clean.iloc[1:-1]
hosp = hosp.reset_index(drop=True)
hosp.head()
```

```python
#Cases DF cleanup
cases_clean = tx_cases_county.iloc[1:,:]
headers = cases_clean.iloc[0].str[-4:]
#Remove excess text & redo heading
for x in range(len(headers)):
    try:
        headers[x] = headers[x]
        headers[x] = months[int(headers[x].split('-')[0])-1] + " " + headers[x].split('-')[1]
    except Exception as e:
        pass

#remove excess data; make headers
cases_clean.iloc[0] = headers
cases_clean.iloc[0,0] = "County"
cases_clean.iloc[0,1] = "Population"
cases_clean.columns = cases_clean.iloc[0]
countypop = cases_clean["Population"]
cases = cases_clean.iloc[1:-11, :]
cases = cases.reset_index(drop=True)
cases = cases.drop(['Population'], axis=1)

#note: rerunning this cell alone will truncate headers to the last 4 characters
cases.head()
```

```python
#County Fatalities DF cleanup
fat_clean = tx_fatality_county.iloc[1:,:]
headers = fat_clean.iloc[0]
#Remove excess text & redo heading
for x in range(len(headers)):
    try:
        headers[x] = headers[x].lstrip('Fatalities ')
        headers[x] = headers[x].lstrip('0')
        headers[x] = months[int(headers[x].split('/')[0])-1] + " " + headers[x].split('/')[1]
    except Exception as e:
        pass
#make new headers and isolate excess rows as another dataframe
fat_clean.iloc[0] = headers
fat_clean.iloc[0,0] = "County"
fat_clean.columns = fat_clean.iloc[0]
countypop = fat_clean["Population"]
totalfat = fat_clean.iloc[-10:-9,:]
fatalities = fat_clean.iloc[1:-10,:]
fatalities = fatalities.reset_index(drop=True)
fatalities = fatalities.drop(['Population'], axis=1)
fatalities.head()
```

```python
#County Tests DF cleanup
#Remove first row
tests_clean = tx_tests_county
headers = tests_clean.iloc[0]
#Remove excess text & redo heading
for x in range(len(headers)):
    try:
        headers[x] = headers[x].lstrip('Tests Through ')
    except Exception as e:
        pass
#make new headers and isolate excess rows as another dataframe
tests_clean.columns = tests_clean.iloc[0]
tests_clean = tests_clean.drop([0])
tests = tests_clean.iloc[0:-10, :]
tests = tests.reset_index(drop=True)
extratests = tests_clean.iloc[-10:, :]
tests.head()
```

```python
#County Active Cases DF cleanup
#Remove first row
activecase_clean = tx_activecases_county.iloc[1:tx_activecases_county.shape[0],1:tx_activecases_county.shape[1]]
#remove excess text occuring for entries before Jun 8, change date format 
active_nobold = activecase_clean.iloc[0].str[-4:]
placeholder = active_nobold
for x in range(len(active_nobold)):
    try:
        active_nobold[x] = active_nobold[x].lstrip('\n')
        active_nobold[x] = months[int(active_nobold[x].split('-')[0])-1] + " " + active_nobold[x].split('-')[1]
    except Exception as e:
        pass

#reinsert headings and drop duplicate row
activecase_clean.iloc[0] = active_nobold
activecase_clean.iloc[0,0] = "County"
activecase_clean.columns = activecase_clean.iloc[0]
activecases = activecase_clean.drop([1])
activecases = activecases.reset_index(drop=True)

#note: rerunning this cell alone will truncate headers to the last 4 characters
activecases.head()
```

# Export Data

```python
tests.to_csv("../resources/data/tests.csv", index=False)
hosp.to_csv("../resources/data/hosp.csv", index=False)
fatalities.to_csv("../resources/data/fatal.csv", index=False)
cases.to_csv("../resources/data/cases.csv", index=False)
activecases.to_csv("../resources/data/actcases.csv", index=False)
```
